```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib.mlab as mlab
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.ticker import LinearLocator, FormatStrFormatter
```

# Pie Chart

```python
labels = ['Penguins', 'Polar bears', 'Lions', 'Otters']
sizes = [30,40,20,10]
colors = ['Green', 'Gray','Red','Blue']

plt.pie (sizes,labels=labels,colors=colors,autopct='%1.1f%%', startangle= 270)
plt.title('Distribution of animals in a zoo')
plt.axis('Equal')
plt.show()
```

#  Bar Chart

```python
objects = ('Comedy', 'Action', 'Romance', 'Drama', 'Sc-fi')
y_pos = np.arange(len(objects))
performance = [10,8,15,2,5]

plt.bar(y_pos, performance, align='center',color= 'green')
plt.xticks(y_pos, objects)
plt.ylabel('number of movies')
plt.title('Movie types')

plt.show()
```

# Grouped Bar Chats 

## side-by-side bar plot

```python
barWidth = 0.40

plt.figure(figsize=(8,5))
children= [780, 1050, 3056, 5025]
adults = [315, 400, 1000, 1500]
 
# Set position of bar on X axis
r1 = np.arange(len(children))
 
# Make the plot
plt.bar(r1, children, color='orange', width=barWidth, label='Children')
plt.bar(r1+barWidth, adults, color='blue', width=barWidth, label='Parents')

 
# Add xticks on the middle of the group bars
plt.xlabel('Visitors', fontweight='bold')
plt.ylabel('Number of visitors')
plt.xticks(r1 + barWidth/2, ['April', 'May', 'June', 'July'])

# Create legend & Show graphic
plt.legend()
plt.show()
```

## stacked bar plot

```python
barWidth = 0.5

plt.figure(figsize=(8,5))

children= [780, 1050, 3056, 5025]
adults = [315, 400, 1000, 1500]

n = np.arange(len(children))
 
plt.bar(n, children, color='red', width=barWidth, label='Children')
plt.bar(n, adults, color='green', width=barWidth, bottom=children, label='Parents')

 
# Add xticks on the middle of the group bars
plt.xlabel('Visitors', fontweight='bold')
plt.ylabel('Number of visitors')
plt.xticks(n, ['April', 'May', 'June', 'July'])

# Create legend & Show graphic
plt.legend()
plt.show()
```

# Line Plot

```python
x = np.arange(-2*np.pi,2*np.pi,np.pi/4)
y = np.sin(x)
z = np.cos(x+np.pi/4)
plt.plot(x,y,marker='o',color='r')
plt.plot(x,z,marker='v',color='g')
plt.xlabel('x values from -2pi to 2pi')
plt.ylabel('sin(x) and cos(x)')
plt.title('Plot of sin and cos from -2pi to 2pi')
plt.legend(['sin(x)', 'cos(x)'], loc = 'best')
plt.show()
```

```python
fig, ax = plt.subplots(1,2,figsize=(8,3))
ax[0].plot(x,y)
ax[0].set_xlabel('x values',fontsize=8)
ax[0].set_ylabel('sin values',fontsize=8)

ax[1].plot(x,z)
ax[1].set_xlabel('x values',fontsize=8)
ax[1].set_ylabel('cos values',fontsize=8)

plt.tight_layout()

plt.show()
```

# Contour Plot

```python
delta = 0.025
x = np.arange(0,4.0,delta)
y = np.arange(0,5.0,delta)

X,Y = np.meshgrid(x,y)
Z = np.exp(-((X-1)**2+(Y-4)**2)/0.15) + \
    np.exp(-((X-3)**2+(Y-4)**2)/0.15) + \
    np.exp(-((X-2)**2+(Y-3)**2)/0.15) + \
    np.exp(-(X-2)**2) * np.exp(-(Y - ((X-2)**2+1))**2/0.15)

plt.figure()
contour_filled = plt.contourf(X,Y,Z,cmap=cm.coolwarm)
 
plt.clabel(contour_filled, colors='k',fmt = '%2.1f',fontsize=12)
plt.colorbar(contour_filled)
plt.title('Filled Contour Plot')
plt.xlabel('X (cm)')
plt.ylabel('Y (cm)')
plt.show()
```

# Surface Plot

```python
fig = plt.figure(figsize=(8,5))
ax = fig.gca(projection='3d') #creating axes

ax.view_init(azim=45) 

# Make data.
X = np.arange(1, 5, 0.025) 
Y = np.arange(1, 5, 0.025) 
X, Y = np.meshgrid(X, Y)
Z = np.exp(-((X-1)**2+(Y-4)**2)/0.15) + \
    np.exp(-((X-3)**2+(Y-4)**2)/0.15) + \
    np.exp(-((X-2)**2+(Y-3)**2)/0.15) + \
    np.exp(-(X-2)**2) * np.exp(-(Y - ((X-2)**2+1))**2/0.15)

surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,linewidth=0, antialiased=False) 

ax.set_zlim(0, 1.01)

ax.zaxis.set_major_locator(LinearLocator(10)) 
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f')) 

fig.colorbar(surf,shrink=0.6,aspect =10,pad=0.09)

ax.set_xlabel('X axis',fontsize=10)
ax.set_ylabel('Y axis',fontsize=10)
ax.set_zlabel('Z axis',fontsize=10)
ax.set_title('Cool-Warm Surface Plot',fontsize=10)
plt.show()
```
