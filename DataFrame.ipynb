```python
import numpy as np
import pandas as pd
```

### How to build a DataFrame

```python
data={'Sex':['Children','Children','Children','Men','Men','Men','Men','Women','Women','Women','Women'],
        'Class':['First','Second','Third','First','Second','Third','Crew','First','Second','Third','Crew'], 
        'Survived':[6,24,27,57,14,75,192,140,80,76,20], 
        'Died':[0,0,52,118,154,387,693,4,13,89,3]}
```

```python
df = pd.DataFrame(data)
df
```

###  .loc 

```python
df.loc[df['Class']=='First']
```

```python
df = df.loc[df['Class']!='Crew']
df
```

### Adding new Coloumn 

```python
df['Total_people'] = df['Survived'] + df['Died']
print(df)
```

```python
df['survived_%'] = df['Survived']/df['Total_people']*100
print(df)
```

### Dropping the coloumn

```python
df = df.drop(['Total_people'],axis=1)
df
```

```python
df.loc[df['survived_%']>80]
```

```python
df.loc[df['survived_%']<40]
```

### GroupBy 

```python
df2 = df.groupby('Class')
_survive = df2['Survived'].sum()
_died = df2['Died'].sum()
print(_survive,"\n")
print(_died,"\n")
total = _survive + _died
print(_survive/total*100,"\n")
print(_died/total*100)
```
